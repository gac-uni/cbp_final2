# NVT Equilibration

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
data_path = "../data/02_processed/00_nvt/"
```

```python
os.system("cat " + data_path + "temperature_header.txt")
```

```python
tem_df = pd.read_csv(data_path + "temperature_raw.xvg", header = None, sep = " ", skipinitialspace=True)

plt.plot(tem_df[0], tem_df[1], label = "instantaneous values")
plt.plot(tem_df[0], tem_df[1].expanding().mean(), label = "cumulative average")
plt.title("temperature in NVT equilibration")
plt.xlabel("time [ps]")
plt.ylabel("temperature [K]")
plt.axhline(y=310, color='r', linestyle='--', linewidth=2, label = "thermostat temperature")
plt.legend()
```
