```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_path = "/home/giacomo/comp_bio/cbp_final/data_other/umbrella/02_processed/wham/"
```

# Basic

```python
comm_path = base_path + "basic_wham/"
hist = np.loadtxt(comm_path+"histo.xvg", comments = ["@", "#"])
r, c = np.shape(hist)

for i in range(c):
    plt.plot(hist[:, 0] * 10, hist[:, i], label = i)

plt.title("Histograms for the Umbrella sampling simulations")
plt.xlabel(r"reaction coordinate $[\AA]$")
plt.ylabel(r"count")
plt.show()

profile = np.loadtxt(comm_path+"profile.xvg", comments = ["@", "#"])

y1 = max(profile[:, 1])
y2 = min(profile[:, 1])
x_start, x_end = 0, 50
label_text = r"$\Delta G \approx$"  f"{y2-y1:.2f} kcal/mol"

fig, ax = plt.subplots()
ax.plot(profile[:, 0] * 10, profile[:, 1])

ax.hlines(y1, x_start, x_end, color='gray', linewidth=1)
ax.hlines(y2, x_start, x_end, color='gray', linewidth=1)
arrow_x = x_end - 3#(x_start + x_end) / 2
ax.annotate('',
            xy=(arrow_x, y1),
            xytext=(arrow_x, y2),
            arrowprops=dict(arrowstyle='<->', color='gray', linewidth=1.5))
ax.text(arrow_x - 10, (y1 + y2) / 2 + 0.05, label_text, color='gray',
        fontsize=12, ha='center')

ax.set_title("Free energy profile of the hDHFR-MTX bind")
ax.set_xlabel(r"distance $[\AA]$")
ax.set_ylabel("Free energy [kcal / mol]")
```

# Convergence

```python
comm_path = "/home/giacomo/comp_bio/cbp_final/data_other/umbrella/02_processed/wham/wham_convergence/"
names = ["p_full", "p_4p5", "p_4p0", "p_3p5", "p_3p0", "p_2p5", "p_2p0"]
deltae = []

for name in names: 
    profile = np.loadtxt(comm_path+name+".xvg", comments = ["@", "#"])
    e_max = np.max(profile[:, 1])
    e_min = np.min(profile[:, 1])
    deltae.append(e_max-e_min)

plt.plot(deltae)
```
