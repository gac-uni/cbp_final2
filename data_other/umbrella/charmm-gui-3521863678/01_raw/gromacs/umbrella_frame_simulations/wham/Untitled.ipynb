```python
profile = np.loadtxt("profile.xvg", comments = ["@", "#"])
```

```python
y1 = max(profile[:, 1])
y2 = min(profile[:, 1])
x_start, x_end = 0, 50
label_text = r"$\Delta G \approx$"  f"{y2-y1:.2f} kcal/mol"

fig, ax = plt.subplots()
ax.plot(profile[:, 0] * 10, profile[:, 1])

ax.hlines(y1, x_start, x_end, color='gray', linewidth=1)
ax.hlines(y2, x_start, x_end, color='gray', linewidth=1)
arrow_x = x_end - 3#(x_start + x_end) / 2
ax.annotate('',
            xy=(arrow_x, y1),
            xytext=(arrow_x, y2),
            arrowprops=dict(arrowstyle='<->', color='gray', linewidth=1.5))
ax.text(arrow_x - 10, (y1 + y2) / 2 + 0.05, label_text, color='gray',
        fontsize=12, ha='center')

ax.set_title("Free energy profile of the hDHFR-MTX bind")
ax.set_xlabel(r"distance $[\AA]$")
ax.set_ylabel("Free energy [kcal / mol]")
```
